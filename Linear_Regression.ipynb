```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Input data: Study Hours (X) and Exam Scores (Y)
X = np.array([2, 4, 6, 8, 10]).reshape(-1, 1)  # Reshaped to a 2D array for sklearn
Y = np.array([50, 65, 80, 90, 95])
```

```python
# Create a Linear Regression model
model = LinearRegression()

# Train the model (fit the line)
model.fit(X, Y)
```

```python
intercept= model.intercept_
intercept
```

```python
# Get the model parameters
intercept = model.intercept_  # beta_0
slope = model.coef_[0]        # beta_1

# Display the equation of the line
print(f"Linear Regression Equation: Y = {intercept:.2f} + {slope:.2f}X")

# Make predictions
Y_pred = model.predict(X)
```

```python
# Calculate Mean Squared Error
mse = mean_squared_error(Y, Y_pred)
print(f"Mean Squared Error: {mse:.2f}")
```

```python
# Plot the data and regression line
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', linestyle='--', label='Regression Line')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```
